```python
import pandas as pd
from homebrewedFunctions.functions import *
df = pd.read_excel("panelDataNoTaxRatios.xlsx", parse_dates = True).rename(columns = {"DATE":"Year"}).set_index(["State", "Year"])
gov_data = pd.read_csv("StateGovFinances2005to2021.csv", parse_dates = True)
gov_data["Year"] = pd.to_datetime(gov_data["Year"])#, format = "%Y")
gov_data.set_index(["State", "Year"], inplace=True)
df.head()
```

```python
gov_data.head()
```

```python
del df["Unnamed: 0"]
df["Revenue"] = gov_data[gov_data["Format"] == "Local government amount"]["1GENERAL REVENUE"]
df["Property Tax"] = gov_data[gov_data["Format"]=="Local government amount"]["1PROPERTY"]
df["Special Assessments"] = gov_data[gov_data["Format"]=="Local government amount"]["1SPECIAL ASSESSMENTS"].str.replace("-", "0").replace("(X)","0").astype(float)
df["Property Tax and Special Assessments"] = df["Property Tax"] + df["Special Assessments"]
df["Sales and Gross Receipts"] = gov_data[gov_data["Format"]=="Local government amount"]["1SALES AND GROSS RECEIPTS"].str.replace("-", "0").replace("(X)","0").astype(float)
df["State-to-Local Transfers"] = gov_data[gov_data["Format"]=="Local government amount"]["1FROM STATE GOVERNMENT"].str.replace("-", "0").replace("(X)","0").astype(float)
df["Federal-to-Local Transfers"] = gov_data[gov_data["Format"]=="Local government amount"]["1FROM FEDERAL GOVERNMENT"]
df["Expenditures"] = gov_data[gov_data["Format"]=="Local government amount"]["1EXPENDITURE"]
df["Deficit"] = df["Expenditures"] - df["Revenue"]
df["Current Charges"] = gov_data[gov_data["Format"]=="Local government amount"]["1CURRENT CHARGES"]#.str.replace("-", "0").replace("(X)","0").astype(float)
df["Revenue - Property Taxes - State-to-Local Transfers"] = df["Revenue"] - df["Property Tax"] - df["State-to-Local Transfers"]
df["Revenue - Property Taxes - Special Assessments - State-to-Local Transfers"] =  df["Revenue - Property Taxes - State-to-Local Transfers"] - df["Special Assessments"]
plot_vars = ["GDP", "Population", "Revenue", "Expenditures", "Property Tax", "Special Assessments", "Property Tax and Special Assessments", "Home Price Index", "Sales and Gross Receipts", "State-to-Local Transfers",
             "Federal-to-Local Transfers",  "Deficit", "Current Charges", 
             "Revenue - Property Taxes - State-to-Local Transfers", 
             "Revenue - Property Taxes - Special Assessments - State-to-Local Transfers"]
```

```python
dfs_dct = {"Level": df}
dfs_dct["Per Capita"] = df.div(df["Population"], axis=0)
dfs_dct["As Percent of Revenue"] = df.div(df["Revenue"], axis=0)
dfs_dct["As Percent of Expenditures"] = df.div(df["Expenditures"], axis=0)
dfs_dct["As Percent of GDP"] = df.div(df["GDP"], axis=0)
for k in list(df.keys())[:12]:
    for dct_key in dfs_dct.keys():
        print(k)
        val = df[k]
        dfs_dct[dct_key][k] = val
dfs_dct = {k: df[plot_vars][df.index.get_level_values("Year") > "2004"] for k, df in dfs_dct.items()}
```
