```python
import pandas as pd
import matplotlib.pyplot as plt
stack_dfs = pd.read_csv("StateGovFinances2009to2021.csv", index_col = ["State", "Year", "Format"]).sort_index()
panel_dfs_dict = {k:stack_dfs[stack_dfs.index.get_level_values(2)==k].reset_index().set_index(["State","Year"]).sort_index() for k in stack_dfs.index.get_level_values(2).unique()}
```

```python
plot_df = panel_dfs_dict['State & local government amount'][["1GENERAL REVENUE","1EXPENDITURE", '1INDIVIDUAL INCOME', '1CORPORATE INCOME', "1PROPERTY", "1SPECIAL ASSESSMENTS"]]
plot_df.rename(columns = {k:k.replace("1", "").title() for k in plot_df.keys()}, inplace = True)
plot_df["Total Income"] = plot_df["Individual Income"].add(plot_df["Corporate Income"])
```

```python
pct_df = plot_df.apply(lambda x: x.div(plot_df["General Revenue"]))
```

```python
states = pct_df.index.get_level_values("State").unique()
color_map = {value: idx for idx, value in enumerate(states)}
pct_df['state'] = pct_df.index.get_level_values('State').map(color_map)
ax = pct_df.plot.scatter(x = "Property", y = "Individual Income", c = "state", colorbar = False, cmap = "viridis", figsize = (20,10))
# # Create a colorbar with the correct labels
cbar = plt.colorbar(ax.collections[0], ticks=range(len(states)))
cbar.ax.set_yticklabels(states)
plt.show()
```

```python
corr_df = pct_df.groupby("State")[["Property", "Total Income"]].corr()["Property"]
corr_df[corr_df.index.get_level_values(1)=="Total Income"].mean()
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
# Define initial columns for x, y, and color
keys =list(pct_df.keys())
x_col = keys[0]
y_col = keys[1]
color_col = keys[2]

# Create the initial scatter plot
fig = px.scatter(pct_df, x=x_col, y=y_col, color=color_col)

# Create the layout for the dropdown menus
dropdown_x = dict(
    buttons=[dict(args=['x', pct_df[col]], label=col, method='restyle') for col in pct_df.columns],
    direction='down',
    showactive=True,
    x=0.17,
    xanchor='left',
    y=1.1,
    yanchor='top'
)

dropdown_y = dict(
    buttons=[dict(args=['y', pct_df[col]], label=col, method='restyle') for col in pct_df.columns],
    direction='down',
    showactive=True,
    x=0.3,
    xanchor='left',
    y=1.1,
    yanchor='top'
)

dropdown_color = dict(
    buttons=[dict(args=['marker.color', pct_df[col]], label=col, method='restyle') for col in pct_df.columns],
    direction='down',
    showactive=True,
    x=0.43,
    xanchor='left',
    y=1.1,
    yanchor='top'
)

# Update the layout of the figure to include the dropdown menus
fig.update_layout(
    updatemenus=[dropdown_x, dropdown_y, dropdown_color],
    title='Interactive Scatter Plot',
    xaxis_title=x_col,
    yaxis_title=y_col
)

# Save the figure to an HTML file
pio.write_html(fig, file='interactive_scatter_plot.html', auto_open=True)
```

```python
from dash import Dash, dcc, html, Input, Output
import selenium
plot_df = pct_df.reset_index()
from selenium import webdriver
import time
import threading
import os

# Function to run the Dash app
def run_dash_app():
    from dash import Dash, dcc, html, Input, Output
    import plotly.express as px
    import pandas as pd

    plot_df = pct_df.reset_index()


    # Initialize the Dash app
    app = Dash(__name__)

    # Define the layout of the app
    app.layout = html.Div([
        html.H1("Interactive Scatter Plot"),
        html.Label("X Axis:"),
        dcc.Dropdown(
            id='x-axis',
            options=[{'label': col, 'value': col} for col in plot_df.columns],
            value=plot_df.columns[0]
        ),
        html.Label("Y Axis:"),
        dcc.Dropdown(
            id='y-axis',
            options=[{'label': col, 'value': col} for col in plot_df.columns],
            value=plot_df.columns[1]
        ),
        html.Label("Color:"),
        dcc.Dropdown(
            id='color',
            options=[{'label': col, 'value': col} for col in plot_df.columns],
            value=plot_df.columns[2]
        ),
        dcc.Graph(id='scatter-plot')
    ])

    # Define the callback to update the scatter plot
    @app.callback(
        Output('scatter-plot', 'figure'),
        Input('x-axis', 'value'),
        Input('y-axis', 'value'),
        Input('color', 'value')
    )
    def update_scatter_plot(x_col, y_col, color_col):
        fig = px.scatter(plot_df, x=x_col, y=y_col, color=color_col)
        return fig

    # Run the app
    app.run_server(debug=False, use_reloader=False)

# Function to save the HTML using Selenium
def save_html():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    driver = webdriver.Chrome(options=options)

    try:
        driver.get("http://127.0.0.1:8050")
        time.sleep(5)  # Wait for the page to load

        with open("interactive_scatter_plot.html", "w") as file:
            file.write(driver.page_source)
    finally:
        driver.quit()

if __name__ == '__main__':
    # Run the Dash app in a separate thread
    app_thread = threading.Thread(target=run_dash_app)
    app_thread.daemon = True
    app_thread.start()

    # Wait for the Dash app to start
    time.sleep(2)

    # Save the HTML
    save_html()

    # Stop the Dash app
    os._exit(0)




# # Initialize the Dash app
# app = Dash(__name__)

# # Define the layout of the app
# app.layout = html.Div([
#     html.H1("Interactive Scatter Plot"),
#     html.Label("X Axis:"),
#     dcc.Dropdown(
#         id='x-axis',
#         options=[{'label': col, 'value': col} for col in plot_df.columns],
#         value=plot_df.columns[0]
#     ),
#     html.Label("Y Axis:"),
#     dcc.Dropdown(
#         id='y-axis',
#         options=[{'label': col, 'value': col} for col in plot_df.columns],
#         value=plot_df.columns[1]
#     ),
#     html.Label("Color:"),
#     dcc.Dropdown(
#         id='color',
#         options=[{'label': col, 'value': col} for col in plot_df.columns],
#         value=plot_df.columns[2]
#     ),
#     dcc.Graph(id='scatter-plot')
# ])

# # Define the callback to update the scatter plot
# @app.callback(
#     Output('scatter-plot', 'figure'),
#     Input('x-axis', 'value'),
#     Input('y-axis', 'value'),
#     Input('color', 'value')
# )
# def update_scatter_plot(x_col, y_col, color_col):
#     fig = px.scatter(plot_df, x=x_col, y=y_col, color=color_col)
#     return fig

# # Run the app
# if __name__ == '__main__':
#     app.run_server(debug=False)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
import threading
import os

# Function to run the Dash app
def run_dash_app():
    from dash import Dash, dcc, html, Input, Output
    import plotly.express as px
    import pandas as pd
    plot_df = pct_df.reset_index()

    # Initialize the Dash app
    app = Dash(__name__)

    # Define the layout of the app
    app.layout = html.Div([
        html.H1("Interactive Scatter Plot"),
        html.Label("X Axis:"),
        dcc.Dropdown(
            id='x-axis',
            options=[{'label': col, 'value': col} for col in plot_df.columns],
            value=plot_df.columns[0]
        ),
        html.Label("Y Axis:"),
        dcc.Dropdown(
            id='y-axis',
            options=[{'label': col, 'value': col} for col in plot_df.columns],
            value=plot_df.columns[1]
        ),
        html.Label("Color:"),
        dcc.Dropdown(
            id='color',
            options=[{'label': col, 'value': col} for col in plot_df.columns],
            value=plot_df.columns[2]
        ),
        dcc.Graph(id='scatter-plot')
    ])

    # Define the callback to update the scatter plot
    @app.callback(
        Output('scatter-plot', 'figure'),
        Input('x-axis', 'value'),
        Input('y-axis', 'value'),
        Input('color', 'value')
    )
    def update_scatter_plot(x_col, y_col, color_col):
        fig = px.scatter(plot_df, x=x_col, y=y_col, color=color_col)
        return fig

    # Run the app
    app.run_server(debug=False, use_reloader=False)

# Function to save the HTML using Selenium
def save_html():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    
    service = Service('/usr/local/bin/chromedriver')
    driver = webdriver.Chrome(service=service, options=options)

    try:
        driver.get("http://127.0.0.1:8050")
        time.sleep(5)  # Wait for the page to load

        with open("interactive_scatter_plot.html", "w") as file:
            file.write(driver.page_source)
    finally:
        driver.quit()

if __name__ == '__main__':
    # Run the Dash app in a separate thread
    app_thread = threading.Thread(target=run_dash_app)
    app_thread.daemon = True
    app_thread.start()

    # Wait for the Dash app to start
    time.sleep(5)

    # Save the HTML
    save_html()

    # Stop the Dash app
    os._exit(0)
```
